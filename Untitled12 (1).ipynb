```python
#initialize 
#grid size
Time = 2 # show increasing 10
L = 10
n = 128
x2 = np.linspace(-L,L,n+1)
x = x2[0:n] #periodicity
dx = x[1] - x[0] #important to calculate correctly
alpha = 2
#time
t = np.linspace(0,2,501)
dt = t[1] - t[0]
time_steps = int(Time/dt)
CFL = alpha*dt/(dx**2)
```

```python
def g_1 (z, lam):
    result = np.abs(1+1/12*lam*(-2*np.cos(2*z)+32*np.cos(z)-30))
    return result
lam = np.arange(0, 3.01, 0.01)
z = np.arange(-np.pi, np.pi+0.1, 0.1)

A1 = (1+1/12*CFL*(-2*np.cos(2*1)+32*np.cos(1)-30))

```

```python
G_one = lambda z: np.abs(g_1(z, CFL)); # Define the function we want to maximize
x = scipy.optimize.fminbound(lambda z:-G_one(z), -np.pi,np.pi) # This gives the *maximizer*
maximum = G_one(x)
print(maximum)
A2 = maximum
```
